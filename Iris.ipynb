```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris
from  sklearn.tree import DecisionTreeClassifier
```

## Wczytanie zbioru z sklearn

```python
data = load_iris()
data['data'].astype(np.float16)
iris = pd.DataFrame(data = np.c_[data['data'], 
                    data.target_names[data.target]],
                    columns = data['feature_names'] + ['target'])
iris [data['feature_names'] ] = iris[ data['feature_names'] ].astype(np.float16)
```

```python
data.feature_names[:]
```

```python
with sns.axes_style (rc={ 'axes.labelcolor': 'black',
                          'figure.facecolor': 'white',
                          'grid.color': '#b0b0b0',
                          'grid.linestyle': '-',
                          'text.color': 'black',
                          'xtick.color': 'black',
                          'ytick.color': 'black'}):
     sns.pairplot(iris, vars=data.feature_names[:4] , hue='target', palette='Set2')
```

## Wczytanie zbioru z seaborn

```python
iris = sns.load_dataset("iris")

with sns.axes_style(rc={ 'axes.labelcolor': 'black',
                         'figure.facecolor': 'white',
                         'grid.color': '#b0b0b0',
                         'grid.linestyle': '-',
                         'text.color': 'black',
                         'xtick.color': 'black',
                         'ytick.color': 'black'}):
    sns.pairplot(iris, hue='species', palette='Set2')
```
